# Euclidean

```python
distances = {}
for other_user in ratings_filled2.index:
    if other_user != 1:
        distance = euclidean(ratings_filled2.loc[1, :], ratings_filled2.loc[other_user, :])
        distances[other_user] = distance
sorted_distances = sorted(distances.items(), key=lambda x: x[1])
print(sorted_distances)
# Calculate the Euclidean distances between User1 and all other users
print(f"The Euclidean distances between User1 and all other users are:")
for user, distance in sorted_distances:
    print(f"{user}: {distance:.2f}")
sorted_distances_df = pd.DataFrame(sorted_distances, columns =['Users', 'Score'])
sorted_distances_df.merge(comparison_df,on="Users",how="inner")
```

# Jaccard similarity

```python
similarities = {}
user =1
for other_user in ratings_filled2.index:
    if other_user != user:
        intersection = ratings_filled2.loc[user, :] & ratings_filled2.loc[other_user, :]
        union = ratings_filled2.loc[user, :] | ratings_filled2.loc[other_user, :]
        similarity = intersection.sum() / union.sum()
        similarities[other_user] = similarity
sorted_similarities = sorted(similarities.items(), key=lambda x: x[1], reverse=True)
print(f"The Jaccard similarity between User1 and all other users, sorted from highest to lowest, are:")
for user, similarity in sorted_similarities:
    print(f"{user}: {similarity:.2f}")
sorted_similarities_df = pd.DataFrame(sorted_similarities, columns =['Users', 'Score'])
sorted_similarities_df.merge(comparison_df,on="Users",how="inner")
```

# Weighted Cosine Similarity

```python
mini_ratings = ratingstest[["DAvg","IAvg","SAvg","CAvg","II","SR","OR","IA","CC","RS"]]
user_test = mini_ratings.iloc[[0]]
conv_arr= user_test.values
user_test = conv_arr.ravel()
result = []
for i in range(len(mini_ratings)):
    user = mini_ratings.iloc[[i]]
    conv_arr= user.values
    user2 = conv_arr.ravel()
    x = cosine(user_test,user2, [DAvg_corr_academics, IAvg_corr_academics, SAvg_corr_academics,CAvg_corr_academics,1,1,1,1,1,1])
    ##df2 = {'DAvg': x[0], 'IAvg': x[1], 'SAvg': x[2],"CAvg":x[3],"II":x[4],"SR":x[5],"OR":x[6],"IA":x[7],"CC":x[8],"RS":x[9]}
    result.append(x)
mini_ratings["WCS"] = result
print(mini_ratings)
```
